```python
# Import Libraries
import ee
import h3
import geopandas as gpd
from tqdm.notebook import tqdm

# Initialize Earth Engine
def initialize_gee(project_id: str, opt_url: str = 'https://earthengine-highvolume.googleapis.com') -> None:
    """Initialize Google Earth Engine with authentication"""
    ee.Authenticate()
    ee.Initialize(project=project_id, opt_url=opt_url)

# Initialize with your project
initialize_gee('ee-nunezrimedio-tesina')  # Replace with your GEE project

# H3 to Earth Engine Conversion Functions
def get_h3_polygon_ee(h3_address):
    """
    Convert H3 address to Earth Engine geometry
    
    Parameters:
    -----------
    h3_address : str
        H3 hexagon address
        
    Returns:
    --------
    ee.Geometry.Polygon
        Earth Engine polygon geometry
    """
    # Get H3 hexagon boundary coordinates as [lat, lon]
    coords = h3.cell_to_boundary(h3_address)

    # Convert to [lon, lat] format required by Earth Engine
    coords_ee = [[lon, lat] for lat, lon in coords]

    # Create and return an Earth Engine polygon
    return ee.Geometry.Polygon([coords_ee])

def create_hexagon_feature(h3_address):
    """
    Create an Earth Engine Feature from an H3 address
    
    Parameters:
    -----------
    h3_address : str
        H3 hexagon address
        
    Returns:
    --------
    ee.Feature
        Earth Engine Feature with geometry and h3_address property
    """
    # Get hexagon geometry
    geometry = get_h3_polygon_ee(h3_address)

    # Create Feature with geometry and add h3_address as property
    return ee.Feature(geometry, {'h3_address': h3_address})

# Simplified NDVI Processing Function - Annual Median Only
def process_ndvi_annual_median(h3_addresses, output_folder='Lithium', batch_name='batch'):
    """
    Process annual median NDVI values for a list of H3 hexagons
    
    Parameters:
    -----------
    h3_addresses : list
        List of H3 hexagon addresses to analyze
    output_folder : str
        Google Drive folder for exporting results
    batch_name : str
        Name identifier for this batch
        
    Returns:
    --------
    ee.batch.Task
        Export task
    """
    # Create EE features from H3 addresses
    features = [create_hexagon_feature(h3_addr) for h3_addr in h3_addresses]
    hexagons = ee.FeatureCollection(features)

    # Get NDVI collection from MODIS
    ndvi_collection = ee.ImageCollection("MODIS/MCD43A4_006_NDVI").select('NDVI')

    def analyze_hexagon(feature):
        geom = feature.geometry()

        # Calculate annual median NDVI for each year
        years = ee.List.sequence(2000, 2022)

        def process_year(year):
            year = ee.Number(year)
            # Convert to integer format without decimal point
            year_str = ee.Number(year).int().format()

            # Define year date range
            year_start = ee.Date.fromYMD(year, 1, 1)
            year_end = ee.Date.fromYMD(year.add(1), 1, 1)

            # Filter NDVI collection for the year
            annual_collection = ndvi_collection.filterDate(year_start, year_end)

            # Calculate median NDVI for the year
            median_ndvi = annual_collection.median()

            # Calculate median value for the hexagon
            annual_median = median_ndvi.reduceRegion(
                reducer=ee.Reducer.median(),
                geometry=geom,
                scale=500,  # MODIS resolution
                bestEffort=True
            )

            # Extract median value or use -9999 if no data
            median_value = ee.Number(ee.Dictionary(annual_median).get('NDVI', -9999))

            # Create property name for this year
            median_key = ee.String('annual_median_').cat(year_str)

            # Return dictionary with year's median
            return ee.Dictionary.fromLists([median_key], [median_value])

        # Process all years and combine results
        yearly_medians = ee.Dictionary(years.map(process_year).iterate(
            lambda dict_year, result: ee.Dictionary(result).combine(ee.Dictionary(dict_year), False),
            ee.Dictionary({})
        ))

        # Add yearly medians as properties to the feature
        return feature.set(yearly_medians)

    # Process all hexagons
    processed_hexagons = hexagons.map(analyze_hexagon)

    # Export to Drive - using a single folder level
    task = ee.batch.Export.table.toDrive(
        collection=processed_hexagons,
        description=f'NDVI_Annual_Median_{batch_name}',
        folder='Lithium_NDVI',  # Single folder name without path separators
        fileFormat='GeoJSON'
    )

    task.start()
    print(f"Task started: NDVI_Annual_Median_{batch_name}")
    return task

# Batch Processing Function
def process_in_batches(h3_addresses, batch_size=500, output_folder='Lithium'):
    """
    Process hexagons in batches to avoid GEE limitations
    
    Parameters:
    -----------
    h3_addresses : list
        List of H3 addresses to analyze
    batch_size : int
        Number of hexagons per batch
    output_folder : str
        Google Drive folder for exporting results
        
    Returns:
    --------
    list
        List of export tasks
    """
    tasks = []
    total_hexagons = len(h3_addresses)

    for i in range(0, total_hexagons, batch_size):
        batch = h3_addresses[i:i+batch_size]
        batch_num = i//batch_size + 1
        total_batches = (total_hexagons+batch_size-1)//batch_size

        print(f"Processing batch {batch_num}/{total_batches} ({len(batch)} hexagons)")
        task = process_ndvi_annual_median(
            batch,
            output_folder=output_folder,
            batch_name=f'batch_{batch_num}_of_{total_batches}'
        )
        tasks.append(task)

    return tasks

# Load Hexagons and Run Analysis
# Load the hexagons file
hexagon_file_path = "/Users/santi/DataspellProjects/Lithium-Jakob/Data/lithium_hexagons_res6.geojson"
gdf = gpd.read_file(hexagon_file_path)
print(f"Loaded {len(gdf)} hexagons")

# Extract H3 addresses
if 'h3_index' in gdf.columns:
    h3_addresses = gdf["h3_index"].to_list()
else:
    h3_addresses = gdf.index.to_list()  # If H3 indices are in the GeoDataFrame index

# Define output folder - single folder name
output_folder = 'Lithium_NDVI'

# Process in batches
tasks = process_in_batches(
    h3_addresses=h3_addresses,
    batch_size=1000,  # Reduced batch size for better reliability
    output_folder=output_folder
)

print(f"Started {len(tasks)} export tasks. Monitor progress in the Earth Engine Code Editor.")
```

```python
# Import Libraries
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from scipy import stats
import os
import glob
import re
import folium
from folium.plugins import MarkerCluster, HeatMap
from branca.element import Template, MacroElement
import branca.colormap as cm
import json
import base64
from io import BytesIO

# Function to calculate trend (slope and R²) from annual NDVI values
def calculate_ndvi_trend(feature_properties):
    """
    Calculate the slope and R² of NDVI change over time using linear regression
    
    Parameters:
    -----------
    feature_properties : dict
        Properties dictionary from a GeoJSON feature containing annual_median_YYYY values
        
    Returns:
    --------
    tuple
        (slope, r_squared, p_value, years, ndvi_values)
    """
    # Extract years and NDVI values
    years = []
    ndvi_values = []

    # Regular expression to match annual_median_YYYY properties
    pattern = r'annual_median_(\d{4})'

    for prop, value in feature_properties.items():
        match = re.match(pattern, prop)
        if match:
            year = int(match.group(1))
            # Skip no-data values (-9999)
            if value != -9999 and value is not None:
                years.append(year)
                ndvi_values.append(float(value))

    # If we have at least 2 points, calculate slope and R²
    if len(years) >= 2:
        # Calculate linear regression
        slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_values)
        r_squared = r_value**2
        return slope, r_squared, p_value, years, ndvi_values
    else:
        # Not enough data points
        return np.nan, np.nan, np.nan, years, ndvi_values

# Function to process a single GeoJSON file and calculate trends
def process_ndvi_file(input_file):
    """
    Process a single GeoJSON file with NDVI data and calculate trends
    
    Parameters:
    -----------
    input_file : str
        Path to GeoJSON file with NDVI data
        
    Returns:
    --------
    geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    """
    print(f"Processing {os.path.basename(input_file)}")

    # Read the GeoJSON file
    gdf = gpd.read_file(input_file)
    print(f"Loaded data: {len(gdf)} hexagons")

    # Calculate trend for each hexagon
    trends = []
    r_squared_values = []
    p_values = []
    time_series_data = []

    for _, feature in gdf.iterrows():
        slope, r_squared, p_value, years, ndvi_values = calculate_ndvi_trend(feature)
        trends.append(slope)
        r_squared_values.append(r_squared)
        p_values.append(p_value)
        time_series_data.append((years, ndvi_values))

    # Add trend as columns
    gdf['ndvi_trend'] = trends
    gdf['r_squared'] = r_squared_values
    gdf['p_value'] = p_values

    # Store time series data for each hexagon (can't directly store in GeoDataFrame)
    gdf['has_time_series'] = [len(x[0]) > 0 for x in time_series_data]

    # To allow access to time series data
    global hexagon_time_series
    hexagon_time_series = {}
    for i, row in gdf.iterrows():
        hexagon_time_series[row['h3_address']] = time_series_data[i]

    # Add trend category for visualization
    def categorize_trend(slope):
        if pd.isna(slope):
            return "insufficient_data"
        elif abs(slope) < 0.0005:  # Threshold for "zero" trend - adjust as needed
            return "stable"
        elif slope > 0:
            return "positive"
        else:
            return "negative"

    gdf['trend_category'] = gdf['ndvi_trend'].apply(categorize_trend)

    # Add significance category
    def categorize_significance(p_value):
        if pd.isna(p_value):
            return "insufficient_data"
        elif p_value <= 0.01:
            return "highly_significant"
        elif p_value <= 0.05:
            return "significant"
        else:
            return "not_significant"

    gdf['significance'] = gdf['p_value'].apply(categorize_significance)

    return gdf

# Function to create a Folium map visualization of NDVI trends
def visualize_ndvi_trends_folium(gdf, output_file=None, title="NDVI Trends (2000-2022)"):
    """
    Create a Folium map visualization of NDVI trends
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output map HTML file
    title : str, optional
        Title for the map
        
    Returns:
    --------
    folium.Map
        Folium Map object with the visualization
    """
    # Get the center of the data for the map
    center = gdf.geometry.unary_union.centroid

    # Create base map
    m = folium.Map(location=[center.y, center.x], zoom_start=6, tiles='CartoDB positron')

    # Add title
    title_html = f'''
         <h3 align="center" style="font-size:16px"><b>{title}</b></h3>
         '''
    m.get_root().html.add_child(folium.Element(title_html))

    # Create a colormap for the trend values
    # Define colors for trends
    trend_colors = {
        "negative": "#cc3333",     # Red
        "stable": "#ffffff",       # White
        "positive": "#33cc33",     # Green
        "insufficient_data": "#cccccc"  # Gray
    }

    # Function to create a mini time series plot for a hexagon
    def create_time_series_plot(h3_address):
        years, ndvi_values = hexagon_time_series[h3_address]

        if len(years) < 2:
            return None

        # Create a figure and plot
        fig, ax = plt.subplots(figsize=(5, 3))
        ax.plot(years, ndvi_values, 'o-', color='blue')

        # Add trend line if we have enough data
        if len(years) >= 2:
            slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_values)
            trend_line = [slope*year + intercept for year in years]
            ax.plot(years, trend_line, 'r--', label=f'Trend: {slope:.6f}/year\nR²: {r_value**2:.3f}')
            ax.legend()

        ax.set_title(f'NDVI Time Series for Hexagon {h3_address[:6]}...')
        ax.set_xlabel('Year')
        ax.set_ylabel('NDVI')
        ax.grid(True, linestyle='--', alpha=0.7)

        # Convert plot to base64 encoded string
        buf = BytesIO()
        plt.tight_layout()
        fig.savefig(buf, format='png', dpi=100)
        plt.close(fig)
        buf.seek(0)
        img_str = base64.b64encode(buf.read()).decode('utf-8')

        return f'<img src="data:image/png;base64,{img_str}">'

    # Add hexagons to the map
    for idx, row in gdf.iterrows():
        # Skip if no geometry
        if row.geometry is None:
            continue

        # Get the trend category and color
        category = row['trend_category']
        color = trend_colors.get(category, "#cccccc")

        # Create tooltip content
        tooltip_content = f"""
        <strong>H3 Address:</strong> {row['h3_address']}<br>
        <strong>Trend:</strong> {row['ndvi_trend']:.6f}<br>
        <strong>R²:</strong> {row['r_squared']:.3f}<br>
        <strong>Category:</strong> {category.capitalize()}<br>
        <strong>Significance:</strong> {'Unknown' if pd.isna(row['p_value']) else f'p-value: {row["p_value"]:.3f}'}
        """

        # Create popup content with time series plot if available
        popup_content = tooltip_content
        if row['has_time_series']:
            time_series_plot = create_time_series_plot(row['h3_address'])
            if time_series_plot:
                popup_content += f"<br>{time_series_plot}"

        # Convert shapely geometry to geojson
        geojson = folium.GeoJson(
            row.geometry.__geo_interface__,
            style_function=lambda x: {
                'fillColor': color,
                'color': '#666666',
                'weight': 0.5,
                'fillOpacity': 0.7
            }
        )

        # Add tooltip
        folium.Tooltip(tooltip_content).add_to(geojson)

        # Add popup with time series if available
        folium.Popup(popup_content, max_width=550).add_to(geojson)

        # Add to map
        geojson.add_to(m)

    # Add a legend
    legend_template = """
    {% macro html(this, kwargs) %}
    <div style="position: fixed; bottom: 50px; right: 50px; width: 180px; z-index:9999; font-size:12px; background-color: white; padding: 10px; border: 2px solid grey; border-radius: 5px;">
        <h4 style="margin-top: 0; text-align: center;">NDVI Trend Legend</h4>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #33cc33; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Positive Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #ffffff; width: 20px; height: 20px; display: inline-block; margin-right: 5px; border: 1px solid #aaa;"></span>
            <span>Stable</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cc3333; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Negative Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cccccc; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Insufficient Data</span>
        </div>
    </div>
    {% endmacro %}
    """

    macro = MacroElement()
    macro._template = Template(legend_template)
    m.get_root().add_child(macro)

    # Add summary statistics to the map
    stats_html = f"""
    <div style="position: fixed; top: 10px; right: 10px; width: 200px; z-index:9998; 
              background-color: white; padding: 10px; border-radius: 5px; border: 1px solid #ccc;">
        <h4 style="margin-top: 0;">Summary Statistics</h4>
        <table style="width: 100%; font-size: 12px;">
            <tr><td>Total Hexagons:</td><td>{len(gdf)}</td></tr>
            <tr><td>Positive Trend:</td><td>{(gdf['trend_category'] == 'positive').sum()} 
                ({(gdf['trend_category'] == 'positive').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Stable:</td><td>{(gdf['trend_category'] == 'stable').sum()} 
                ({(gdf['trend_category'] == 'stable').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Negative Trend:</td><td>{(gdf['trend_category'] == 'negative').sum()} 
                ({(gdf['trend_category'] == 'negative').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Median NDVI Trend:</td><td>{gdf['ndvi_trend'].median():.6f}</td></tr>
            <tr><td>Mean R²:</td><td>{gdf['r_squared'].mean():.3f}</td></tr>
        </table>
    </div>
    """
    m.get_root().html.add_child(folium.Element(stats_html))

    # Save map if output file specified (commented out for testing)
    # if output_file:
    #     m.save(output_file)
    #     print(f"Saved map to {output_file}")

    return m

# Function to create a comprehensive NDVI analysis dashboard
def create_ndvi_dashboard(gdf, output_file=None):
    """
    Create a detailed NDVI analysis dashboard with various visualizations
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output HTML file
        
    Returns:
    --------
    str
        Path to saved HTML file
    """
    # Create the main folium map
    m = visualize_ndvi_trends_folium(gdf, title="NDVI Trends in Lithium Areas (2000-2022)")

    # Save dashboard to HTML (commented out for testing)
    # if output_file:
    #     m.save(output_file)
    #     print(f"Saved dashboard to {output_file}")

    return m

# Function to generate a detailed analysis report
def generate_trend_analysis_table(gdf, output_file=None):
    """
    Generate a detailed table of NDVI trend analysis results
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output CSV file
        
    Returns:
    --------
    pandas.DataFrame
        DataFrame with trend analysis results
    """
    # Select relevant columns
    report_df = gdf[['h3_address', 'ndvi_trend', 'r_squared', 'p_value', 'trend_category', 'significance']]

    # Add significance level description
    def get_significance_description(p_value):
        if pd.isna(p_value):
            return "Insufficient data"
        elif p_value <= 0.01:
            return "Highly significant (p ≤ 0.01)"
        elif p_value <= 0.05:
            return "Significant (p ≤ 0.05)"
        else:
            return f"Not significant (p = {p_value:.3f})"

    report_df['significance_desc'] = report_df['p_value'].apply(get_significance_description)

    # Save to CSV if output file specified (commented out for testing)
    # if output_file:
    #     report_df.to_csv(output_file, index=False)
    #     print(f"Saved trend analysis table to {output_file}")

    return report_df

# Function to create various NDVI time series plots
def create_ndvi_time_series_plots(gdf, output_folder=None):
    """
    Create various NDVI time series plots
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_folder : str, optional
        Folder to save the output plots
        
    Returns:
    --------
    dict
        Dictionary of matplotlib figures
    """
    figures = {}

    # Create output folder if specified (commented out for testing)
    # if output_folder and not os.path.exists(output_folder):
    #     os.makedirs(output_folder)

    # 1. Overall NDVI trend plot (average across all hexagons by year)
    # First, collect all annual NDVI values
    annual_data = {}

    for h3_address, (years, values) in hexagon_time_series.items():
        for year, value in zip(years, values):
            if year not in annual_data:
                annual_data[year] = []
            annual_data[year].append(value)

    # Calculate statistics for each year
    years = sorted(annual_data.keys())
    ndvi_means = [np.mean(annual_data[year]) for year in years]
    ndvi_medians = [np.median(annual_data[year]) for year in years]
    ndvi_std = [np.std(annual_data[year]) for year in years]

    # Create overall trend plot
    fig1, ax1 = plt.subplots(figsize=(10, 6))
    ax1.errorbar(years, ndvi_means, yerr=ndvi_std, fmt='o-', capsize=5,
                 label='Mean NDVI ± Std Dev', color='blue')
    ax1.plot(years, ndvi_medians, 's--', color='red', label='Median NDVI')

    # Add trend line
    if len(years) >= 2:
        slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_means)
        trend_line = [slope*year + intercept for year in years]
        ax1.plot(years, trend_line, 'g-', linewidth=2,
                 label=f'Trend: {slope:.6f}/year (R²={r_value**2:.3f})')

    ax1.set_title('Overall NDVI Trend (2000-2022)', fontsize=14)
    ax1.set_xlabel('Year', fontsize=12)
    ax1.set_ylabel('NDVI', fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend()
    plt.tight_layout()

    figures['overall_trend'] = fig1

    # if output_folder:
    #     fig1.savefig(os.path.join(output_folder, 'overall_ndvi_trend.png'), dpi=300)

    # 2. Distribution of NDVI trends
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    trends = gdf['ndvi_trend'].dropna()
    ax2.hist(trends, bins=30, alpha=0.7, color='skyblue', edgecolor='black')

    # Add vertical line at median
    median_trend = trends.median()
    ax2.axvline(median_trend, color='red', linestyle='--',
                label=f'Median: {median_trend:.6f}')

    # Add vertical line at zero
    ax2.axvline(0, color='black', linestyle='-')

    ax2.set_title('Distribution of NDVI Trends', fontsize=14)
    ax2.set_xlabel('NDVI Trend (change/year)', fontsize=12)
    ax2.set_ylabel('Frequency', fontsize=12)
    ax2.legend()
    plt.tight_layout()

    figures['trend_distribution'] = fig2

    # if output_folder:
    #     fig2.savefig(os.path.join(output_folder, 'ndvi_trend_distribution.png'), dpi=300)

    # 3. Boxplot of NDVI values by year
    fig3, ax3 = plt.subplots(figsize=(12, 6))

    # Prepare data for boxplot
    box_data = []
    box_labels = []

    for year in years:
        box_data.append(annual_data[year])
        box_labels.append(str(year))

    ax3.boxplot(box_data, labels=box_labels)
    ax3.set_title('NDVI Distribution by Year', fontsize=14)
    ax3.set_xlabel('Year', fontsize=12)
    ax3.set_ylabel('NDVI', fontsize=12)
    ax3.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Rotate x labels for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()

    figures['annual_boxplot'] = fig3

    # if output_folder:
    #     fig3.savefig(os.path.join(output_folder, 'ndvi_annual_boxplot.png'), dpi=300)

    # 4. Heat map of R² vs Trend
    fig4, ax4 = plt.subplots(figsize=(8, 6))

    # Get data
    trends = gdf['ndvi_trend'].dropna()
    r2_values = gdf['r_squared'].dropna()

    # Only plot where both values exist
    mask = ~np.isnan(trends) & ~np.isnan(r2_values)

    if mask.sum() > 0:
        scatter = ax4.scatter(trends[mask], r2_values[mask],
                              c=gdf['p_value'][mask], cmap='viridis_r',
                              alpha=0.7, edgecolors='black', linewidth=0.5)

        # Add colorbar
        cbar = plt.colorbar(scatter)
        cbar.set_label('p-value', rotation=270, labelpad=15)

        # Add horizontal line at p=0.05
        ax4.axhline(0.05, color='red', linestyle='--', label='p=0.05')

        ax4.set_title('R² vs NDVI Trend', fontsize=14)
        ax4.set_xlabel('NDVI Trend (change/year)', fontsize=12)
        ax4.set_ylabel('R²', fontsize=12)
        ax4.grid(True, linestyle='--', alpha=0.7)
        ax4.legend()
    else:
        ax4.text(0.5, 0.5, "Insufficient data for this plot",
                 ha='center', va='center', transform=ax4.transAxes)

    plt.tight_layout()

    figures['r2_vs_trend'] = fig4

    # if output_folder:
    #     fig4.savefig(os.path.join(output_folder, 'r2_vs_trend.png'), dpi=300)

    return figures

# Set the path to your GeoJSON file with NDVI data
input_file = "/Users/santi/DataspellProjects/Lithium-Jakob/Data/NDVI_Annual_Median_batch_1_of_1.geojson"  # Change to your file path

# Process NDVI file and calculate trends
ndvi_data = process_ndvi_file(input_file)

# Display basic information about the data
print(f"NDVI data processed: {len(ndvi_data)} hexagons analyzed")
print(f"Trend categories: {ndvi_data['trend_category'].value_counts().to_dict()}")

# Generate detailed trend analysis table (without saving to file)
trend_table = generate_trend_analysis_table(ndvi_data)

# Display the first few rows of the trend analysis table
display(trend_table.head())

# Create Folium map visualization (without saving to file)
ndvi_map = create_ndvi_dashboard(ndvi_data)

# Display the map in the notebook
ndvi_map

# Create different visualizations of NDVI trends over time (without saving to files)
time_series_plots = create_ndvi_time_series_plots(ndvi_data)

# Display the plots
for plot_name, fig in time_series_plots.items():
    print(f"\n{plot_name.replace('_', ' ').title()}:")
    display(fig)
```

```python
# Function to create a Folium map visualization with CUENCA information
def visualize_ndvi_trends_folium_with_cuenca(gdf, title="NDVI Trends by Basin (2000-2022)"):
    """
    Create a Folium map visualization of NDVI trends with CUENCA filtering
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons, trend data, and CUENCA information
    title : str, optional
        Title for the map
        
    Returns:
    --------
    folium.Map
        Folium Map object with the visualization
    """
    # Get the center of the data for the map
    center = gdf.geometry.unary_union.centroid

    # Create base map
    m = folium.Map(location=[center.y, center.x], zoom_start=6, tiles='CartoDB positron')

    # Add title
    title_html = f'''
         <h3 align="center" style="font-size:16px"><b>{title}</b></h3>
         '''
    m.get_root().html.add_child(folium.Element(title_html))

    # Define colors for trends
    trend_colors = {
        "negative": "#cc3333",     # Red
        "stable": "#ffffff",       # White
        "positive": "#33cc33",     # Green
        "insufficient_data": "#cccccc"  # Gray
    }

    # Create layer groups for each CUENCA (for filtering)
    cuenca_layers = {}

    # Get unique CUENCAs
    unique_cuencas = gdf['primary_cuenca'].unique()

    # Create a layer group for each CUENCA
    for cuenca in unique_cuencas:
        cuenca_layers[cuenca] = folium.FeatureGroup(name=f"Basin: {cuenca}")
        m.add_child(cuenca_layers[cuenca])

    # Function to create a mini time series plot for a hexagon
    def create_time_series_plot(h3_address):
        years, ndvi_values = hexagon_time_series[h3_address]

        if len(years) < 2:
            return None

        # Create a figure and plot
        fig, ax = plt.subplots(figsize=(5, 3))
        ax.plot(years, ndvi_values, 'o-', color='blue')

        # Add trend line if we have enough data
        if len(years) >= 2:
            slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_values)
            trend_line = [slope*year + intercept for year in years]
            ax.plot(years, trend_line, 'r--', label=f'Trend: {slope:.6f}/year\nR²: {r_value**2:.3f}')
            ax.legend()

        ax.set_title(f'NDVI Time Series for Hexagon {h3_address[:6]}...')
        ax.set_xlabel('Year')
        ax.set_ylabel('NDVI')
        ax.grid(True, linestyle='--', alpha=0.7)

        # Convert plot to base64 encoded string
        buf = BytesIO()
        plt.tight_layout()
        fig.savefig(buf, format='png', dpi=100)
        plt.close(fig)
        buf.seek(0)
        img_str = base64.b64encode(buf.read()).decode('utf-8')

        return f'<img src="data:image/png;base64,{img_str}">'

    # Add hexagons to the appropriate CUENCA layer
    for idx, row in gdf.iterrows():
        # Skip if no geometry
        if row.geometry is None:
            continue

        # Get the trend category and color
        category = row['trend_category']
        color = trend_colors.get(category, "#cccccc")

        # Get CUENCA
        cuenca = row['primary_cuenca']

        # Create tooltip content
        tooltip_content = f"""
        <strong>H3 Address:</strong> {row['h3_address']}<br>
        <strong>Basin:</strong> {cuenca}<br>
        <strong>Trend:</strong> {row['ndvi_trend']:.6f}<br>
        <strong>R²:</strong> {row['r_squared']:.3f}<br>
        <strong>Category:</strong> {category.capitalize()}<br>
        <strong>Significance:</strong> {'Unknown' if pd.isna(row['p_value']) else f'p-value: {row["p_value"]:.3f}'}
        """

        # Create popup content with time series plot if available
        popup_content = tooltip_content
        if row['has_time_series']:
            time_series_plot = create_time_series_plot(row['h3_address'])
            if time_series_plot:
                popup_content += f"<br>{time_series_plot}"

        # Create a style dictionary directly with fixed colors
        style = {
            'fillColor': color,
            'color': '#666666',
            'weight': 0.5,
            'fillOpacity': 0.7
        }

        # Convert shapely geometry to geojson and apply fixed style
        geojson = folium.GeoJson(
            row.geometry.__geo_interface__,
            style_function=lambda x, style=style: style,
            tooltip=folium.Tooltip(tooltip_content),
            popup=folium.Popup(popup_content, max_width=550)
        )

        # Add to the appropriate CUENCA layer
        geojson.add_to(cuenca_layers[cuenca])

    # Add a legend
    legend_template = """
    {% macro html(this, kwargs) %}
    <div style="position: fixed; bottom: 50px; right: 50px; width: 180px; z-index:9999; font-size:12px; background-color: white; padding: 10px; border: 2px solid grey; border-radius: 5px;">
        <h4 style="margin-top: 0; text-align: center;">NDVI Trend Legend</h4>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #33cc33; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Positive Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #ffffff; width: 20px; height: 20px; display: inline-block; margin-right: 5px; border: 1px solid #aaa;"></span>
            <span>Stable</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cc3333; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Negative Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cccccc; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Insufficient Data</span>
        </div>
    </div>
    {% endmacro %}
    """

    macro = MacroElement()
    macro._template = Template(legend_template)
    m.get_root().add_child(macro)

    # Add summary statistics to the map
    stats_html = f"""
    <div style="position: fixed; top: 10px; right: 10px; width: 200px; z-index:9998; 
              background-color: white; padding: 10px; border-radius: 5px; border: 1px solid #ccc;">
        <h4 style="margin-top: 0;">Summary Statistics</h4>
        <table style="width: 100%; font-size: 12px;">
            <tr><td>Total Hexagons:</td><td>{len(gdf)}</td></tr>
            <tr><td>Basins:</td><td>{len(unique_cuencas)}</td></tr>
            <tr><td>Positive Trend:</td><td>{(gdf['trend_category'] == 'positive').sum()} 
                ({(gdf['trend_category'] == 'positive').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Stable:</td><td>{(gdf['trend_category'] == 'stable').sum()} 
                ({(gdf['trend_category'] == 'stable').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Negative Trend:</td><td>{(gdf['trend_category'] == 'negative').sum()} 
                ({(gdf['trend_category'] == 'negative').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Median NDVI Trend:</td><td>{gdf['ndvi_trend'].median():.6f}</td></tr>
            <tr><td>Mean R²:</td><td>{gdf['r_squared'].mean():.3f}</td></tr>
        </table>
    </div>
    """
    m.get_root().html.add_child(folium.Element(stats_html))

    # Add layer control to toggle basin layers
    folium.LayerControl().add_to(m)

    return m# Function to visualize NDVI trends by CUENCA
def visualize_ndvi_by_cuenca(gdf, cuenca_analysis):
    """
    Create visualizations to compare NDVI trends between different CUENCAs
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons, trend data, and CUENCA information
    cuenca_analysis : dict
        Dictionary with CUENCA analysis results from analyze_ndvi_by_cuenca
        
    Returns:
    --------
    dict
        Dictionary with matplotlib figures for different visualizations
    """
    figures = {}

    # Extract CUENCA statistics and time series data
    cuenca_stats = cuenca_analysis['cuenca_stats']
    cuenca_time_series = cuenca_analysis['cuenca_time_series']

    # 1. Bar chart comparing mean NDVI trend by CUENCA
    fig1, ax1 = plt.subplots(figsize=(12, 6))

    # Get top CUENCAs by hexagon count (to avoid plotting very small CUENCAs)
    top_cuencas = cuenca_stats.sort_values('hexagon_count', ascending=False).head(10)['primary_cuenca'].tolist()

    # Filter stats for top CUENCAs
    plot_stats = cuenca_stats[cuenca_stats['primary_cuenca'].isin(top_cuencas)]

    # Sort by mean NDVI trend
    plot_stats = plot_stats.sort_values('ndvi_trend_mean')

    # Create color map based on trend direction
    colors = ['#cc3333' if x < 0 else '#33cc33' for x in plot_stats['ndvi_trend_mean']]

    # Plot bar chart
    bars = ax1.barh(plot_stats['primary_cuenca'], plot_stats['ndvi_trend_mean'], color=colors)

    # Add error bars
    ax1.errorbar(
        plot_stats['ndvi_trend_mean'],
        plot_stats['primary_cuenca'],
        xerr=plot_stats['ndvi_trend_std'],
        fmt='none',
        color='black',
        capsize=5
    )

    # Add vertical line at zero
    ax1.axvline(0, color='black', linestyle='--', alpha=0.7)

    # Add labels
    for i, bar in enumerate(bars):
        width = bar.get_width()
        label_x_pos = width + 0.0001 if width >= 0 else width - 0.0003
        ax1.text(label_x_pos,
                 bar.get_y() + bar.get_height()/2,
                 f"{width:.6f}",
                 va='center',
                 fontsize=9)

    ax1.set_title('Mean NDVI Trend by Basin (2000-2022)', fontsize=14)
    ax1.set_xlabel('NDVI Trend (change/year)', fontsize=12)
    ax1.set_ylabel('Basin (CUENCA)', fontsize=12)
    ax1.grid(axis='x', linestyle='--', alpha=0.7)

    figures['cuenca_trends_bar'] = fig1

    # 2. Stacked bar chart of trend categories by CUENCA
    fig2, ax2 = plt.subplots(figsize=(12, 6))

    # Filter for top CUENCAs
    trend_data = cuenca_stats[cuenca_stats['primary_cuenca'].isin(top_cuencas)].copy()

    # Sort by percentage of positive trends
    trend_data = trend_data.sort_values('positive')

    # Create stacked bar chart
    trend_data.plot(
        x='primary_cuenca',
        y=['positive', 'stable', 'negative'],
        kind='barh',
        stacked=True,
        ax=ax2,
        color=['#33cc33', '#eeeeee', '#cc3333'],
        width=0.8
    )

    # Add labels with percentages
    for container in ax2.containers:
        ax2.bar_label(container, fmt='%.1f%%', label_type='center')

    ax2.set_title('Distribution of NDVI Trend Categories by Basin', fontsize=14)
    ax2.set_xlabel('Percentage of Hexagons', fontsize=12)
    ax2.set_ylabel('Basin (CUENCA)', fontsize=12)
    ax2.grid(axis='x', linestyle='--', alpha=0.3)
    ax2.legend(['Positive', 'Stable', 'Negative'], loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

    plt.tight_layout()
    figures['cuenca_trend_categories'] = fig2

    # 3. Time series plot comparing NDVI trends across CUENCAs
    fig3, ax3 = plt.subplots(figsize=(12, 8))

    # Get top 5 CUENCAs by hexagon count for time series plotting
    top5_cuencas = cuenca_stats.sort_values('hexagon_count', ascending=False).head(5)['primary_cuenca'].tolist()

    # Plot time series for each CUENCA
    for i, cuenca in enumerate(top5_cuencas):
        if cuenca not in cuenca_time_series:
            continue

        ts_data = cuenca_time_series[cuenca]
        years = ts_data['years']
        means = ts_data['means']
        stds = ts_data['stds']
        trend_line = ts_data['trend_line']

        # Plot with error bands
        ax3.plot(years, means, 'o-', linewidth=2, label=f"{cuenca} (slope={ts_data['slope']:.6f})")
        ax3.fill_between(years,
                         [m - s for m, s in zip(means, stds)],
                         [m + s for m, s in zip(means, stds)],
                         alpha=0.2)

        # Plot trend line
        ax3.plot(years, trend_line, '--', linewidth=1, alpha=0.7)

    ax3.set_title('NDVI Trends by Basin (2000-2022)', fontsize=14)
    ax3.set_xlabel('Year', fontsize=12)
    ax3.set_ylabel('Mean NDVI', fontsize=12)
    ax3.grid(True, linestyle='--', alpha=0.7)
    ax3.legend()

    plt.tight_layout()
    figures['cuenca_time_series'] = fig3

    # 4. Boxplot comparing NDVI trends distribution across CUENCAs
    fig4, ax4 = plt.subplots(figsize=(12, 6))

    # Prepare data for boxplot
    boxplot_data = []
    boxplot_labels = []

    for cuenca in top5_cuencas:
        # Get all NDVI trend values for this CUENCA
        trends = gdf[gdf['primary_cuenca'] == cuenca]['ndvi_trend'].dropna().tolist()

        if trends:
            boxplot_data.append(trends)
            # Add count for reference
            boxplot_labels.append(f"{cuenca}\n(n={len(trends)})")

    # Create boxplot
    ax4.boxplot(boxplot_data, labels=boxplot_labels, patch_artist=True)

    # Add horizontal line at zero
    ax4.axhline(0, color='black', linestyle='--', alpha=0.7)

    ax4.set_title('Distribution of NDVI Trends by Basin', fontsize=14)
    ax4.set_ylabel('NDVI Trend (change/year)', fontsize=12)
    ax4.set_xlabel('Basin (CUENCA)', fontsize=12)
    ax4.grid(axis='y', linestyle='--', alpha=0.7)

    # Rotate x labels
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    figures['cuenca_trend_boxplot'] = fig4

    return figures# Function to analyze NDVI trends by CUENCA
def analyze_ndvi_by_cuenca(gdf):
    """
    Analyze NDVI trends grouped by CUENCA
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons, trend data, and CUENCA information
        
    Returns:
    --------
    dict
        Dictionary with various CUENCA analysis results
    """
    # Group by CUENCA and calculate statistics
    cuenca_stats = gdf.groupby('primary_cuenca').agg({
        'ndvi_trend': ['mean', 'median', 'std', 'count'],
        'r_squared': ['mean', 'median'],
        'h3_address': 'count'  # Count of hexagons
    })

    # Flatten multi-index columns
    cuenca_stats.columns = ['_'.join(col).strip() for col in cuenca_stats.columns.values]
    cuenca_stats.reset_index(inplace=True)

    # Rename columns for clarity
    cuenca_stats.rename(columns={'h3_address_count': 'hexagon_count'}, inplace=True)

    # Calculate percentage of positive, negative, and stable trends for each CUENCA
    trend_percentages = {}

    for cuenca in gdf['primary_cuenca'].unique():
        cuenca_subset = gdf[gdf['primary_cuenca'] == cuenca]
        total = len(cuenca_subset)

        if total == 0:
            continue

        trend_counts = cuenca_subset['trend_category'].value_counts()
        percentages = {
            'positive': trend_counts.get('positive', 0) / total * 100,
            'stable': trend_counts.get('stable', 0) / total * 100,
            'negative': trend_counts.get('negative', 0) / total * 100,
            'insufficient_data': trend_counts.get('insufficient_data', 0) / total * 100
        }

        trend_percentages[cuenca] = percentages

    # Create a DataFrame with trend percentages
    trend_df = pd.DataFrame(trend_percentages).T.reset_index()
    trend_df.rename(columns={'index': 'primary_cuenca'}, inplace=True)

    # Join with cuenca_stats
    cuenca_stats = cuenca_stats.merge(trend_df, on='primary_cuenca', how='left')

    # Sort by mean NDVI trend
    cuenca_stats.sort_values('ndvi_trend_mean', ascending=False, inplace=True)

    # Calculate time series data by CUENCA
    cuenca_time_series = {}

    for cuenca in gdf['primary_cuenca'].unique():
        cuenca_subset = gdf[gdf['primary_cuenca'] == cuenca]

        # For each year, collect all NDVI values from this CUENCA
        annual_data = {}

        for _, row in cuenca_subset.iterrows():
            h3_address = row['h3_address']
            if h3_address in hexagon_time_series:
                years, values = hexagon_time_series[h3_address]

                for year, value in zip(years, values):
                    if year not in annual_data:
                        annual_data[year] = []
                    annual_data[year].append(value)

        # Calculate annual statistics for this CUENCA
        years = sorted(annual_data.keys())
        if not years:
            continue

        means = [np.mean(annual_data[year]) for year in years]
        medians = [np.median(annual_data[year]) for year in years]
        stds = [np.std(annual_data[year]) for year in years]

        # Calculate trend for the means
        if len(years) >= 2:
            slope, intercept, r_value, p_value, std_err = stats.linregress(years, means)
            trend_line = [slope*year + intercept for year in years]

            cuenca_time_series[cuenca] = {
                'years': years,
                'means': means,
                'medians': medians,
                'stds': stds,
                'slope': slope,
                'r_squared': r_value**2,
                'p_value': p_value,
                'trend_line': trend_line
            }

    return {
        'cuenca_stats': cuenca_stats,
        'cuenca_time_series': cuenca_time_series
    }# Import Libraries
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from scipy import stats
import os
import glob
import re
import folium
from folium.plugins import MarkerCluster, HeatMap
from branca.element import Template, MacroElement
import branca.colormap as cm
import json
import base64
from io import BytesIO
import seaborn as sns

# Function to calculate trend (slope and R²) from annual NDVI values
def calculate_ndvi_trend(feature_properties):
    """
    Calculate the slope and R² of NDVI change over time using linear regression
    
    Parameters:
    -----------
    feature_properties : dict
        Properties dictionary from a GeoJSON feature containing annual_median_YYYY values
        
    Returns:
    --------
    tuple
        (slope, r_squared, p_value, years, ndvi_values)
    """
    # Extract years and NDVI values
    years = []
    ndvi_values = []

    # Regular expression to match annual_median_YYYY properties
    pattern = r'annual_median_(\d{4})'

    for prop, value in feature_properties.items():
        match = re.match(pattern, prop)
        if match:
            year = int(match.group(1))
            # Skip no-data values (-9999)
            if value != -9999 and value is not None:
                years.append(year)
                ndvi_values.append(float(value))

    # If we have at least 2 points, calculate slope and R²
    if len(years) >= 2:
        # Calculate linear regression
        slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_values)
        r_squared = r_value**2
        return slope, r_squared, p_value, years, ndvi_values
    else:
        # Not enough data points
        return np.nan, np.nan, np.nan, years, ndvi_values

# Function to process a single GeoJSON file and calculate trends, with CUENCA data join
def process_ndvi_file_with_cuenca(ndvi_file, hexagons_file):
    """
    Process NDVI data file and join with hexagons file containing CUENCA information
    
    Parameters:
    -----------
    ndvi_file : str
        Path to GeoJSON file with NDVI data
    hexagons_file : str
        Path to GeoJSON file with CUENCA information
        
    Returns:
    --------
    geopandas.GeoDataFrame
        GeoDataFrame with hexagons, trend data, and CUENCA information
    """
    print(f"Processing NDVI data: {os.path.basename(ndvi_file)}")

    # Read the NDVI GeoJSON file
    ndvi_gdf = gpd.read_file(ndvi_file)
    print(f"Loaded NDVI data: {len(ndvi_gdf)} hexagons")

    # Read the hexagons file with CUENCA information
    hexagons_gdf = gpd.read_file(hexagons_file)
    print(f"Loaded hexagons data: {len(hexagons_gdf)} hexagons")

    # Verify column names in both datasets
    print(f"NDVI columns: {ndvi_gdf.columns.tolist()}")
    print(f"Hexagons columns: {hexagons_gdf.columns.tolist()}")

    # Determine the joining keys
    ndvi_id_col = 'h3_address' if 'h3_address' in ndvi_gdf.columns else None
    hexagon_id_col = 'h3_index' if 'h3_index' in hexagons_gdf.columns else None

    if not ndvi_id_col:
        raise ValueError("Could not find h3_address column in NDVI data")

    if not hexagon_id_col:
        raise ValueError("Could not find h3_index column in hexagons data")

    print(f"Joining on NDVI column '{ndvi_id_col}' and hexagons column '{hexagon_id_col}'")

    # Calculate trend for each NDVI hexagon
    trends = []
    r_squared_values = []
    p_values = []
    time_series_data = []

    for _, feature in ndvi_gdf.iterrows():
        slope, r_squared, p_value, years, ndvi_values = calculate_ndvi_trend(feature)
        trends.append(slope)
        r_squared_values.append(r_squared)
        p_values.append(p_value)
        time_series_data.append((years, ndvi_values))

    # Add trend information to the NDVI GeoDataFrame
    ndvi_gdf['ndvi_trend'] = trends
    ndvi_gdf['r_squared'] = r_squared_values
    ndvi_gdf['p_value'] = p_values

    # Store time series data for each hexagon
    ndvi_gdf['has_time_series'] = [len(x[0]) > 0 for x in time_series_data]

    # Global variable to store time series data
    global hexagon_time_series
    hexagon_time_series = {}
    for i, row in ndvi_gdf.iterrows():
        hexagon_time_series[row[ndvi_id_col]] = time_series_data[i]

    # Add trend category for visualization
    def categorize_trend(slope):
        if pd.isna(slope):
            return "insufficient_data"
        elif abs(slope) < 0.0005:  # Threshold for "zero" trend - adjust as needed
            return "stable"
        elif slope > 0:
            return "positive"
        else:
            return "negative"

    ndvi_gdf['trend_category'] = ndvi_gdf['ndvi_trend'].apply(categorize_trend)

    # Add significance category
    def categorize_significance(p_value):
        if pd.isna(p_value):
            return "insufficient_data"
        elif p_value <= 0.01:
            return "highly_significant"
        elif p_value <= 0.05:
            return "significant"
        else:
            return "not_significant"

    ndvi_gdf['significance'] = ndvi_gdf['p_value'].apply(categorize_significance)

    # Prepare for join with hexagons data
    # Make a copy of the joining column with the same name for merging
    ndvi_gdf[hexagon_id_col] = ndvi_gdf[ndvi_id_col]

    # Drop geometry before merge to avoid conflicts
    hexagons_no_geom = hexagons_gdf.drop(columns='geometry')

    # Join the datasets
    merged_gdf = ndvi_gdf.merge(hexagons_no_geom, on=hexagon_id_col, how='left')

    # Check how many hexagons have CUENCA information
    cuenca_count = merged_gdf['CUENCA'].notna().sum()
    print(f"Successfully joined CUENCA information for {cuenca_count} of {len(merged_gdf)} hexagons")

    # Handle potential missing CUENCA values
    merged_gdf['CUENCA'].fillna('Unknown', inplace=True)

    # Split multiple basins (if any have semicolon-separated values)
    def get_primary_cuenca(cuenca_str):
        if ';' in str(cuenca_str):
            return cuenca_str.split(';')[0]
        return cuenca_str

    merged_gdf['primary_cuenca'] = merged_gdf['CUENCA'].apply(get_primary_cuenca)

    return merged_gdf

# Function to create a Folium map visualization of NDVI trends
def visualize_ndvi_trends_folium(gdf, output_file=None, title="NDVI Trends (2000-2022)"):
    """
    Create a Folium map visualization of NDVI trends
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output map HTML file
    title : str, optional
        Title for the map
        
    Returns:
    --------
    folium.Map
        Folium Map object with the visualization
    """
    # Get the center of the data for the map
    center = gdf.geometry.unary_union.centroid

    # Create base map
    m = folium.Map(location=[center.y, center.x], zoom_start=6, tiles='CartoDB positron')

    # Add title
    title_html = f'''
         <h3 align="center" style="font-size:16px"><b>{title}</b></h3>
         '''
    m.get_root().html.add_child(folium.Element(title_html))

    # Create a colormap for the trend values
    # Define colors for trends
    trend_colors = {
        "negative": "#cc3333",     # Red
        "stable": "#ffffff",       # White
        "positive": "#33cc33",     # Green
        "insufficient_data": "#cccccc"  # Gray
    }

    # Function to create a mini time series plot for a hexagon
    def create_time_series_plot(h3_address):
        years, ndvi_values = hexagon_time_series[h3_address]

        if len(years) < 2:
            return None

        # Create a figure and plot
        fig, ax = plt.subplots(figsize=(5, 3))
        ax.plot(years, ndvi_values, 'o-', color='blue')

        # Add trend line if we have enough data
        if len(years) >= 2:
            slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_values)
            trend_line = [slope*year + intercept for year in years]
            ax.plot(years, trend_line, 'r--', label=f'Trend: {slope:.6f}/year\nR²: {r_value**2:.3f}')
            ax.legend()

        ax.set_title(f'NDVI Time Series for Hexagon {h3_address[:6]}...')
        ax.set_xlabel('Year')
        ax.set_ylabel('NDVI')
        ax.grid(True, linestyle='--', alpha=0.7)

        # Convert plot to base64 encoded string
        buf = BytesIO()
        plt.tight_layout()
        fig.savefig(buf, format='png', dpi=100)
        plt.close(fig)
        buf.seek(0)
        img_str = base64.b64encode(buf.read()).decode('utf-8')

        return f'<img src="data:image/png;base64,{img_str}">'

    # Add hexagons to the map
    for idx, row in gdf.iterrows():
        # Skip if no geometry
        if row.geometry is None:
            continue

        # Get the trend category and color
        category = row['trend_category']
        color = trend_colors.get(category, "#cccccc")

        # Create tooltip content
        tooltip_content = f"""
        <strong>H3 Address:</strong> {row['h3_address']}<br>
        <strong>Trend:</strong> {row['ndvi_trend']:.6f}<br>
        <strong>R²:</strong> {row['r_squared']:.3f}<br>
        <strong>Category:</strong> {category.capitalize()}<br>
        <strong>Significance:</strong> {'Unknown' if pd.isna(row['p_value']) else f'p-value: {row["p_value"]:.3f}'}
        """

        # Create popup content with time series plot if available
        popup_content = tooltip_content
        if row['has_time_series']:
            time_series_plot = create_time_series_plot(row['h3_address'])
            if time_series_plot:
                popup_content += f"<br>{time_series_plot}"

        # Create a style dictionary directly with fixed colors
        style = {
            'fillColor': color,
            'color': '#666666',
            'weight': 0.5,
            'fillOpacity': 0.7
        }

        # Convert shapely geometry to geojson and apply fixed style
        folium.GeoJson(
            row.geometry.__geo_interface__,
            style_function=lambda x, style=style: style,
            tooltip=folium.Tooltip(tooltip_content),
            popup=folium.Popup(popup_content, max_width=550)
        ).add_to(m)

    # Add a legend
    legend_template = """
    {% macro html(this, kwargs) %}
    <div style="position: fixed; bottom: 50px; right: 50px; width: 180px; z-index:9999; font-size:12px; background-color: white; padding: 10px; border: 2px solid grey; border-radius: 5px;">
        <h4 style="margin-top: 0; text-align: center;">NDVI Trend Legend</h4>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #33cc33; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Positive Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #ffffff; width: 20px; height: 20px; display: inline-block; margin-right: 5px; border: 1px solid #aaa;"></span>
            <span>Stable</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cc3333; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Negative Trend</span>
        </div>
        <div style="display: flex; align-items: center; margin: 5px 0;">
            <span style="background-color: #cccccc; width: 20px; height: 20px; display: inline-block; margin-right: 5px;"></span>
            <span>Insufficient Data</span>
        </div>
    </div>
    {% endmacro %}
    """

    macro = MacroElement()
    macro._template = Template(legend_template)
    m.get_root().add_child(macro)

    # Add summary statistics to the map
    stats_html = f"""
    <div style="position: fixed; top: 10px; right: 10px; width: 200px; z-index:9998; 
              background-color: white; padding: 10px; border-radius: 5px; border: 1px solid #ccc;">
        <h4 style="margin-top: 0;">Summary Statistics</h4>
        <table style="width: 100%; font-size: 12px;">
            <tr><td>Total Hexagons:</td><td>{len(gdf)}</td></tr>
            <tr><td>Positive Trend:</td><td>{(gdf['trend_category'] == 'positive').sum()} 
                ({(gdf['trend_category'] == 'positive').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Stable:</td><td>{(gdf['trend_category'] == 'stable').sum()} 
                ({(gdf['trend_category'] == 'stable').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Negative Trend:</td><td>{(gdf['trend_category'] == 'negative').sum()} 
                ({(gdf['trend_category'] == 'negative').sum() / len(gdf) * 100:.1f}%)</td></tr>
            <tr><td>Median NDVI Trend:</td><td>{gdf['ndvi_trend'].median():.6f}</td></tr>
            <tr><td>Mean R²:</td><td>{gdf['r_squared'].mean():.3f}</td></tr>
        </table>
    </div>
    """
    m.get_root().html.add_child(folium.Element(stats_html))

    # Save map if output file specified (commented out for testing)
    # if output_file:
    #     m.save(output_file)
    #     print(f"Saved map to {output_file}")

    return m

# Function to create a comprehensive NDVI analysis dashboard
def create_ndvi_dashboard(gdf, output_file=None):
    """
    Create a detailed NDVI analysis dashboard with various visualizations
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output HTML file
        
    Returns:
    --------
    str
        Path to saved HTML file
    """
    # Create the main folium map
    m = visualize_ndvi_trends_folium(gdf, title="NDVI Trends in Lithium Areas (2000-2022)")

    # Save dashboard to HTML (commented out for testing)
    # if output_file:
    #     m.save(output_file)
    #     print(f"Saved dashboard to {output_file}")

    return m

# Function to generate a detailed analysis report
def generate_trend_analysis_table(gdf, output_file=None):
    """
    Generate a detailed table of NDVI trend analysis results
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_file : str, optional
        Path to save the output CSV file
        
    Returns:
    --------
    pandas.DataFrame
        DataFrame with trend analysis results
    """
    # Select relevant columns
    report_df = gdf[['h3_address', 'ndvi_trend', 'r_squared', 'p_value', 'trend_category', 'significance']]

    # Add significance level description
    def get_significance_description(p_value):
        if pd.isna(p_value):
            return "Insufficient data"
        elif p_value <= 0.01:
            return "Highly significant (p ≤ 0.01)"
        elif p_value <= 0.05:
            return "Significant (p ≤ 0.05)"
        else:
            return f"Not significant (p = {p_value:.3f})"

    report_df['significance_desc'] = report_df['p_value'].apply(get_significance_description)

    # Save to CSV if output file specified (commented out for testing)
    # if output_file:
    #     report_df.to_csv(output_file, index=False)
    #     print(f"Saved trend analysis table to {output_file}")

    return report_df

# Function to create various NDVI time series plots
def create_ndvi_time_series_plots(gdf, output_folder=None):
    """
    Create various NDVI time series plots
    
    Parameters:
    -----------
    gdf : geopandas.GeoDataFrame
        GeoDataFrame with hexagons and trend data
    output_folder : str, optional
        Folder to save the output plots
        
    Returns:
    --------
    dict
        Dictionary of matplotlib figures
    """
    figures = {}

    # Create output folder if specified (commented out for testing)
    # if output_folder and not os.path.exists(output_folder):
    #     os.makedirs(output_folder)

    # 1. Overall NDVI trend plot (average across all hexagons by year)
    # First, collect all annual NDVI values
    annual_data = {}

    for h3_address, (years, values) in hexagon_time_series.items():
        for year, value in zip(years, values):
            if year not in annual_data:
                annual_data[year] = []
            annual_data[year].append(value)

    # Calculate statistics for each year
    years = sorted(annual_data.keys())
    ndvi_means = [np.mean(annual_data[year]) for year in years]
    ndvi_medians = [np.median(annual_data[year]) for year in years]
    ndvi_std = [np.std(annual_data[year]) for year in years]

    # Create overall trend plot
    fig1, ax1 = plt.subplots(figsize=(10, 6))
    ax1.errorbar(years, ndvi_means, yerr=ndvi_std, fmt='o-', capsize=5,
                 label='Mean NDVI ± Std Dev', color='blue')
    ax1.plot(years, ndvi_medians, 's--', color='red', label='Median NDVI')

    # Add trend line
    if len(years) >= 2:
        slope, intercept, r_value, p_value, std_err = stats.linregress(years, ndvi_means)
        trend_line = [slope*year + intercept for year in years]
        ax1.plot(years, trend_line, 'g-', linewidth=2,
                 label=f'Trend: {slope:.6f}/year (R²={r_value**2:.3f})')

    ax1.set_title('Overall NDVI Trend (2000-2022)', fontsize=14)
    ax1.set_xlabel('Year', fontsize=12)
    ax1.set_ylabel('NDVI', fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend()
    plt.tight_layout()

    figures['overall_trend'] = fig1

    # if output_folder:
    #     fig1.savefig(os.path.join(output_folder, 'overall_ndvi_trend.png'), dpi=300)

    # 2. Distribution of NDVI trends
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    trends = gdf['ndvi_trend'].dropna()
    ax2.hist(trends, bins=30, alpha=0.7, color='skyblue', edgecolor='black')

    # Add vertical line at median
    median_trend = trends.median()
    ax2.axvline(median_trend, color='red', linestyle='--',
                label=f'Median: {median_trend:.6f}')

    # Add vertical line at zero
    ax2.axvline(0, color='black', linestyle='-')

    ax2.set_title('Distribution of NDVI Trends', fontsize=14)
    ax2.set_xlabel('NDVI Trend (change/year)', fontsize=12)
    ax2.set_ylabel('Frequency', fontsize=12)
    ax2.legend()
    plt.tight_layout()

    figures['trend_distribution'] = fig2

    # if output_folder:
    #     fig2.savefig(os.path.join(output_folder, 'ndvi_trend_distribution.png'), dpi=300)

    # 3. Boxplot of NDVI values by year
    fig3, ax3 = plt.subplots(figsize=(12, 6))

    # Prepare data for boxplot
    box_data = []
    box_labels = []

    for year in years:
        box_data.append(annual_data[year])
        box_labels.append(str(year))

    ax3.boxplot(box_data, labels=box_labels)
    ax3.set_title('NDVI Distribution by Year', fontsize=14)
    ax3.set_xlabel('Year', fontsize=12)
    ax3.set_ylabel('NDVI', fontsize=12)
    ax3.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Rotate x labels for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()

    figures['annual_boxplot'] = fig3

    # if output_folder:
    #     fig3.savefig(os.path.join(output_folder, 'ndvi_annual_boxplot.png'), dpi=300)

    # 4. Heat map of R² vs Trend
    fig4, ax4 = plt.subplots(figsize=(8, 6))

    # Get data
    trends = gdf['ndvi_trend'].dropna()
    r2_values = gdf['r_squared'].dropna()

    # Only plot where both values exist
    mask = ~np.isnan(trends) & ~np.isnan(r2_values)

    if mask.sum() > 0:
        scatter = ax4.scatter(trends[mask], r2_values[mask],
                              c=gdf['p_value'][mask], cmap='viridis_r',
                              alpha=0.7, edgecolors='black', linewidth=0.5)

        # Add colorbar
        cbar = plt.colorbar(scatter)
        cbar.set_label('p-value', rotation=270, labelpad=15)

        # Add horizontal line at p=0.05
        ax4.axhline(0.05, color='red', linestyle='--', label='p=0.05')

        ax4.set_title('R² vs NDVI Trend', fontsize=14)
        ax4.set_xlabel('NDVI Trend (change/year)', fontsize=12)
        ax4.set_ylabel('R²', fontsize=12)
        ax4.grid(True, linestyle='--', alpha=0.7)
        ax4.legend()
    else:
        ax4.text(0.5, 0.5, "Insufficient data for this plot",
                 ha='center', va='center', transform=ax4.transAxes)

    plt.tight_layout()

    figures['r2_vs_trend'] = fig4

    # if output_folder:
    #     fig4.savefig(os.path.join(output_folder, 'r2_vs_trend.png'), dpi=300)

    return figures

# Set the paths to your GeoJSON files
ndvi_file = "/Users/santi/DataspellProjects/Lithium-Jakob/Data/NDVI_Annual_Median_batch_1_of_1.geojson"
hexagons_file = "/Users/santi/DataspellProjects/Lithium-Jakob/Data/lithium_hexagons_res6.geojson"

# Process NDVI file and join with CUENCA information
ndvi_data = process_ndvi_file_with_cuenca(ndvi_file, hexagons_file)

# Display basic information about the data
print(f"NDVI data processed: {len(ndvi_data)} hexagons analyzed")
print(f"Trend categories: {ndvi_data['trend_category'].value_counts().to_dict()}")
print(f"Basins (CUENCAs): {ndvi_data['primary_cuenca'].nunique()}")

# Analyze NDVI trends by CUENCA
cuenca_analysis = analyze_ndvi_by_cuenca(ndvi_data)

# Display CUENCA statistics
display(cuenca_analysis['cuenca_stats'].head(10))

# Generate detailed trend analysis table (without saving to file)
trend_table = generate_trend_analysis_table(ndvi_data)

# Display the first few rows of the trend analysis table
display(trend_table.head())

# Create visualizations comparing NDVI trends between CUENCAs
cuenca_viz = visualize_ndvi_by_cuenca(ndvi_data, cuenca_analysis)

# Display the CUENCA comparison visualizations
for plot_name, fig in cuenca_viz.items():
    print(f"\n{plot_name.replace('_', ' ').title()}:")
    display(fig)

# Create Folium map visualization with CUENCA layers (without saving to file)
ndvi_map = visualize_ndvi_trends_folium_with_cuenca(ndvi_data)

# Display the map in the notebook
ndvi_map
```
